## Examples of Data Distributions

### Uniform Distribution
<p>Flat constante probability of a value ocorrecy in a given arrange.</p>

```python
import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

### Normal / Gaussian
<p>Visualize the probability density function:</p>

```python
from scipy.stats import norm

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
```

```python
nu = 5.0
sigma_standard_desviation = 2.0
values = np.random.normal(nu, sigma_standard_desviation, 10000)
plt.hist(values, 50)
plt.show()
```

#### Exponential PDF(Probability Distribution Function) / "Power Law"

```python
from scipy.stats import expon

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
```

#### Binomial Probability Mass Function

```python
from scipy.stats import binom

n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

#### Poisson Probability Mass Function
<p>Ex: My website gets on average 500 visits per day.</p>
<p>What's the odds of getting 550?</p>

```python
from scipy.stats import poisson

mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
```
