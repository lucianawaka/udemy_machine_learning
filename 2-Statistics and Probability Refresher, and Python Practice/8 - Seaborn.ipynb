## Seaborn
Seaborn is a visalization library that sits on top of matplotlib, making it nicer to look at and adding some extra capabilities too.  
Let's start by loading up a real dataset on 2019 model-year vehicles, and plotting a histogram just using matplotlib to see the distribution by number of gears.

```python
# imports
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading data and exploring data
df = pd.read_csv("http://media.sundog-soft.com/SelfDriving/FuelEfficiency.csv")
df.head()
```

```python
# Plot bar - Gears
gear_counts = df['# Gears'].value_counts()
gear_counts.plot(kind='bar');
```

## Seaborn
Make matplotlib look better

```python
import seaborn as sns
sns.set() 
# Change matplotlib's default settings to look more visually pleasing
```

```python
gear_counts.plot(kind='bar');
```

### Histplot
#### Replicating basic distplot output
Both new functions support the kernel density estimate line, by passing <b>kde=True</b>:

```python
sns.histplot(df['CombMPG'], kde=True);
```

### Pairplot

```python
df.columns
```

```python
df2 = df[['Cylinders','CityMPG', 'HwyMPG', 'CombMPG']]
```

```python
df2.head()
```

```python
sns.pairplot(df2, hue='Cylinders', height=2.5);
```

### Scatterplot

```python
sns.scatterplot(x="Eng Displ", y="CombMPG", data=df);
```

### Jointplot

```python
sns.jointplot(x="Eng Displ", y="CombMPG", data=df);
```

### Lmplot
Is a scatterplot 
With a linear regression line computed and overlaid onto the data

```python
sns.lmplot(x="Eng Displ", y="CombMPG", data=df);
```

### Boxplot
From seaborn

```python
sns.set(rc={'figure.figsize':(18, 6)})
ax=sns.boxplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

### Stripplot

```python
ax=sns.stripplot(x='Mfr Name', y='CombMPG', data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

### Countplot

```python
ax=sns.countplot(x='Mfr Name', data=df)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

#### Numerical Values => Histogram
#### Categorical Values => Countplot

### Heatmap

```python
df2 = df.pivot_table(index='Cylinders', columns='Eng Displ', values='CombMPG', aggfunc='mean')
sns.heatmap(df2);
```
