### Linear Regression


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3
```

```python
plt.scatter(pageSpeeds, purchaseAmount);
plt.xlabel('Page Speed (s)')
plt.ylabel('Purchase Amount')
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
```

```python
# Not surprisngly, our R-squared value shows a really good fit
r_value ** 2
```

### The slope 
Indicates the steepness of a line (Indica a inclinação de uma linha)
### The intercept 
Indicates the location where it intersects an axis (Indica o local onde ela intercepta um eixo)

```python
# The slope is negative - 3.007
# The intercept is 100.009
# Formula => y = -3.007x + 100.009
```

```python
def predict(x):
    return slope * x + intercept
```

```python
fitLine = predict(pageSpeeds)
```

```python
# Plotting the Regression Line
plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='g')
plt.show()
```
