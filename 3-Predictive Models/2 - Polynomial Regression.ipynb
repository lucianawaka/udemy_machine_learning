### Polynomial Regression

```python
# Libaries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
np.random.seed(2)
pageSpeed = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeed
```

```python
dict = {'pageSpeed':pageSpeed, 'purchaseAmount':purchaseAmount}
```

```python
df = pd.DataFrame(dict)
```

```python
sns.scatterplot(data =df,x = pageSpeed, y = purchaseAmount);
```

```python
## 4th degree Polynomial for this data
```

```python
x = df['pageSpeed'] #np.array(pageSpeed)
y = df['purchaseAmount'] #np.array(purchaseAmount)

p4 = np.poly1d(np.polyfit(x, y, 4))
```

```python
xp = np.linspace(0, 7, 100)
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y, p4(x))
print(r2)
```
